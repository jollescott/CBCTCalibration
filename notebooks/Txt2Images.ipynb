```python
for i in range(0,4):
    
    file_object  = open('/data/data1/BB_data/BB_simulated_tomographies/open_beam_BB/{0}/sample_camera.dat'.format(i), 'read')
  
    myfile = file_object.read()
    id_start = myfile.find('I:')+3
    id_end = myfile.find('# Error')-1
    mytxtimag = myfile[id_start:id_end]
    file_object.close()
    splitted = mytxtimag.split()
    myimg = np.zeros((1024,1024))

    v=0;

    for row in range(0,1024):
        for col in range(0,1024):
    #     print(pixel)
            myimg[row][col] = round(float(splitted[v]))
#             print(np.uint16(float(splitted[v])+0.5))
            v=v+1
    
    myi = myimg.astype('uint16')        

    io.imsave('/data/data1/BB_data/BB_simulated_tomographies/openbeams_BB/ob{:04d}.tiff'.format(i), myi)
```
